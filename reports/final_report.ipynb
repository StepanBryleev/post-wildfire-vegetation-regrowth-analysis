## 2016 Chimney Tops 2 Fire

In mid-November 2016, a forest fire started on Chimney Tops Mountain in Great Smoky Mountains National Park. Due to unseasonably dry and windy conditions, the fire grew several orders of magnitude in one day and spread uncontrollably throughout the national park towards the nearby town of Gatlinburg. [The Chimney Top 2 fire](https://www.nps.gov/grsm/learn/chimney-tops-2-fire.htm) caused major destruction that impacted homes, businesses and also [resulted in fatalities and injuries](https://edition.cnn.com/2016/12/05/us/tennessee-gatlinburg-wildfires/index.html). The Chimney Tops 2 wildfire was recorded as the largest fire in Great Smoky Mountains National Park history and is of great interest to a wide range of environmental researchers.

## Remote sensing as the research method 

All around the world wildfires are becoming more and more severe from year to year. So there is a greater need to study evolution of wildfires and post fire changes for maintaining resources, protecting ecosystems, people and properties. One such way is to use remote sensing technologies to study fires from the air or from space. Remote sensing data can provide sufficient land area coverage and frequent measurements to study fires efficiently and to better understand them.

During the Chimney Tops 2 Fire (CT2) over 10,000 acres of Great Smoky Mountains National Park burned including the area in which two NEON field sites are located. Thus, detailed remote sensing analysis can be used to identify vegetation regrowth dynamics inside the fire perimeter and evaluate post-fire ecological changes.

## Data source selection

NEON began data collection activities at LeConte and Great Smoky Mountains in 2015, when the [Airborne Observation Platform (AOP)](https://www.neonscience.org/data-collection/airborne-remote-sensing) flew over both sites collecting LiDAR, hyperspectral and high resolution RGB ortho-photos of the surrounding vegetation and landscape. The sites were flown again in the summer of 2016 at peak greenness, and the fall of 2017 past peak greenness. So, we are fortunate to have a little over one year's worth of pre-fire data and subsequent data on fire recovery. This allows to complete detailed analysis of the vegetation regrowth inside Chimney Top 2 fire perimeter.

Although the NEON spectrometer has high spatial resolution and is a great instrument, the reflectance data is typically only collected by NEON once a year. Thus, more frequent Landsat data with lower spatial resolution can be used to validate NEON data and provide a more comprehensive time series. 

## Study area

To estimate post-fire changes and recovery rates study reigion of 36.35 km<sup>2</sup> was dfined inside the CT2 fire perimeter. Unfortunately, due to weather and other operational reasons coverage area of NEON airborne observations in GRSM differ from year to year. To avoid "NO DATA" gaps, study polygon was selected that excludes the presence of "NO DATA" pixels both for NEON reflectance and for NEON Lidar data.

Initially, Canopy Height Models for 2016, 2017 and 2021 observations were calculated to estimate landscape changes. 

Then we compared interannual AOP reflectance data measured by the NEON Imaging Spectrometer (NIS) in 2016, 2017 and 2021 with interannual data from the Operational Land Imager on the Landsat 8 satellite.

Although the NEON spectrometer has high spatial resolution (1-meter), the NEON reflectance data is typically only available once a year. Moreover, in 2017 GRSM site was flown in fall of 2017 which is past peak greenness. So it is difficult to compose representative time series for estimating regrowth dynamis rates using only NIS data.

Since Landsat 8 orbits the Earth every 16 days, there is more reflectance data available. Thus, more frequent but lower spatial resolution (30-meter) Landsat 8 data was used to validate NEON data and provide a more comprehensive time series. 

All analysis was implemented in [Google Earth Engine](https://earthengine.google.com/).
